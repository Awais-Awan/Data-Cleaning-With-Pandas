```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel(r"Customer Call List.xlsx")
df=pd.DataFrame(data)
df
```

```python
df.duplicated().sum()
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns="Not_Useful_Column")
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip("...")
df["Last_Name"]=df["Last_Name"].str.strip("/")
df["Last_Name"]=df["Last_Name"].str.strip("_")
df
```

```python
#or we can simply do
df["Last_Name"]=df["Last_Name"].str.strip("123./_")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace(r'\D', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].replace(['NaN', ''], np.nan)
df
```

```python
# df['Phone_Number']=df['Phone_Number'].apply(lambda num: str(num)[0:3] + "-" + str(num)[3:6]+ "-" + str(num)[6:10])
# df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace("nan--","")
df
```

```python
df["Address"].str.split(',',n=2,expand=True)
```

```python
df[["Street","State","Zip_Code"]]=df["Address"].str.split(',',n=2,expand=True)
df
```

```python
def fun(x):
    if x=="Yes":
        x="Yes"
        return x
    elif x=="Y":
        x="Yes"
        return x
    if x=="No":
        x="No"
        return x
    elif x=="N":
        x="No"
        return x
    
df["Paying Customer"]=df["Paying Customer"].map(fun)
df["Do_Not_Contact"]=df["Do_Not_Contact"].map(fun)
df
```

```python
df=df.replace("N/a","")
# df.replace("NaN","")#we cant do this it will not work
df=df.fillna("")
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Yes':
        df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)
df
#or we can also do
# df=df.dropna(subset="Phone_Number"),inplace=True
```
